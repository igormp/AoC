```python
import heapq
from collections import Counter
from math import prod
import numpy as np
```

```python
#FILE = "example.txt"
FILE = "input.txt"
```

```python
inputs = [
    list(map(int, line.rstrip('\n').split(','))) for line in open(FILE)
]
```

```python
distances = np.zeros((len(inputs), len(inputs)))
for i in range(len(inputs)):
    for j in range(len(inputs)):
        a, b = inputs[i], inputs[j]
        distances[i, j] = np.linalg.norm(np.array(a) - np.array(b))
```

```python
# Gather all unique point pairs (i, j) with their distance, ignoring i == j
# Only consider each unordered pair once: i < j
pair_distances = []

for i in range(len(inputs)):
    for j in range(i + 1, len(inputs)):
        pair_distances.append((i, inputs[i], j, inputs[j], distances[i, j]))

# Sort the list by distance
pair_distances.sort(key=lambda x: x[4])

# Print as "point X - point Y: distance Z" from closest to furthest
for i, pt1, j, pt2, dist in pair_distances[:20]:
    print(f"Point {i} at {pt1} - Point {j} at {pt2}: {dist:.6f}")
```

```python
len(pair_distances)
```

```python
groups: list[set[int]] = []
remaining = 10
limit = 1000
for i, pt1, j, pt2, dist in pair_distances[:limit]:
    if len(groups) == 0:
        #print(f"adding {i} and {j} to new group")
        groups.append({i,j})
        remaining -= 1
        continue

    groups_added = 0
    for group in groups:
        if i in group or j in group:
           # print(f"adding {i} and {j} to {group}")
            group.add(i)
            group.add(j)
            groups_added += 1
    if groups_added == 0:
        #print(f"adding {i} and {j} to new group")
        groups.append({i,j})
        remaining -= 1

    remaining -= groups_added
    
    if groups_added > 1:
        merged_set = set()
        idxs = []
        for idx, group in enumerate(groups):
            if i in group or j in group:
                merged_set |= group
                idxs.append(idx)
        for idx in sorted(idxs, reverse=True):
            groups.pop(idx)
        # print(f"merged {idxs} into {merged_set}")
        groups.append(merged_set)

    
    # if remaining <= 0:
    #     print("done")
    #     break
```
