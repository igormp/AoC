```python
import heapq
from collections import Counter, defaultdict, deque
from itertools import product
from math import prod
import numpy as np
from dataclasses import dataclass
from functools import lru_cache
```

```python
#FILE = "example2.txt"
FILE = "input.txt"
```

```python
inputs = [
    line.rstrip('\n').split(' ') for line in open(FILE)
]
```

```python
parsed_inputs = {
    input[0].strip(':'): set(input[1:])
    for input in inputs
}
```

```python
inverted_index = defaultdict(set)
for k, v in parsed_inputs.items():
    for v_item in v:
        inverted_index[v_item].add(k)
```

```python
@lru_cache(maxsize=None)
def count_paths(node, target):
    if node == target:
        return 1
    total = 0
    for next_item in parsed_inputs.get(node, []):
        total += count_paths(next_item, target)
    return total

results = count_paths('svr', 'fft') * count_paths('fft', 'dac') * count_paths('dac', 'out')
results += count_paths('svr', 'dac') * count_paths('dac', 'fft') * count_paths('fft', 'out')
```
