```python
#FILE = "example.txt"
FILE = "input.txt"
```

```python
inputs = [
    list(line.split()) for line in open(FILE)
]
```

```python
ranges = []
while inputs:
    line = inputs.pop(0)
    if not line:
        break
    ranges.extend(line)
ranges = [(int(i.split('-')[0]),int(i.split('-')[1])) for i in ranges]
```

```python
ranges = sorted(sorted(ranges, key=lambda x: x[1]), key=lambda x: x[0])
```

```python
# ranges = [(0,2),(1,10),(12,20),(13,15),(13,16),(25,30),(25,26),(26,26),(31,31),(31,32),(31,33),(32,40),(39,41),(45,45),(45,45)]
# ranges = sorted(sorted(ranges, key=lambda x: x[1]), key=lambda x: x[0])
idx = 0
end_ranges = []
while idx < len(ranges):
    cur_range = ranges[idx]
    i = idx + 1
    end_candidate = cur_range[1]
    while i < len(ranges) and (
        (ranges[i][0] <= end_candidate or ranges[i][1] <= end_candidate)
    ):
        end_candidate = max(ranges[i][1],end_candidate)
        i += 1
    end_ranges.append((cur_range[0], end_candidate))
    idx = i
```

```python
fresh_ids = 0
for start,end in end_ranges:
    fresh_ids += end-start+1
```
