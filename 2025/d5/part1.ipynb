```python
#FILE = "example.txt"
FILE = "input.txt"
```

```python
inputs = [
    list(line.split()) for line in open(FILE)
]
```

```python
inputs
```

```python
ranges = []
while inputs:
    line = inputs.pop(0)
    if not line:
        break
    ranges.extend(line)
ranges = [(int(i.split('-')[0]),int(i.split('-')[1])) for i in ranges]
```

```python
ranges = sorted(ranges, key=lambda x: x[0])
#ranges
```

```python
len(ranges)
```

```python
numbers = [int(i[0]) for i in inputs]
```

```python
numbers.sort()
```

```python
fresh_count = 0
freshes = []
# for number in numbers:
#     for start,end in ranges:
#         if start <= number <= end:
#             break
#     else:
#         spoiled_count += 1
for number in numbers:
    spoiled = True
    for start,end in ranges:
        if start <= number <= end:
            #print(f"{number} is fresh at range {start}-{end}")
            freshes.append(number)
            spoiled = False
            break
    if not spoiled:
        fresh_count += 1
```
