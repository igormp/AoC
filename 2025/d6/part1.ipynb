```python
import heapq
from collections import Counter
from math import prod
```

```python
#FILE = "example.txt"
FILE = "input.txt"
```

```python
inputs = [
    list(line.split()) for line in open(FILE)
]
```

```python
inputs = list(zip(*inputs))
inputs = [list(i) for i in inputs]
```

```python
inputs
```

```python
for idigit, i in enumerate(inputs):
    i = [int(j) for j in i if j.isdigit()] + [i[-1]]
    inputs[idigit] = i
```

```python
inputs
```

```python
total = 0
for i in inputs:
    res = 0
    match i[-1]:
        case '+':
            res = sum(i[:-1])
        case '*':
            res = prod(i[:-1])
    total += res
```
