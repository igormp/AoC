```python
start = inputs[0].index('S')
```

```python
r, c = len(inputs), len(inputs[0])

mem = [[0] * c for _ in range(r)]
mem[0][start] = 1


for i in range(r-1):
    for j in range(c):
        possibilities = mem[i][j]
        if possibilities == 0:
            continue

        cur = inputs[i+1][j]
        if cur == '^':
            mem[i+1][j-1] += possibilities
            mem[i+1][j+1] += possibilities
        else:
            mem[i+1][j] += possibilities
```
