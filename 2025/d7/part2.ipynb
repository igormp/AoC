```python
start = inputs[0].index('S')
```

```python
r, c = len(inputs), len(inputs[0])

possibilities = [(1, start)]

total_paths = 0

while possibilities:
    r, c = possibilities.pop()
    if r == len(inputs) - 1:
        total_paths += 1
        continue

    cur = inputs[r][c]
    if cur == '^':
        new_paths = [(r+1, c-1), (r+1, c+1)]
    else:
        new_paths = [(r+1, c)]

    for new_r, new_c in new_paths:
        possibilities.append((new_r, new_c))
```
