```python
#FILE = "example.txt"
FILE = "input.txt"
```

```python
inputs = [
    list(line.rstrip('\n')) for line in open(FILE)
]
```

```python
start = inputs[0].index('S')
```

```python
set_count = set([start])
print(''.join(inputs[0]))
splits = 0
for i in inputs[1:]:
    if '^' not in i:
        line = i.copy()
        #print(f"aa {set_count}")
        #print(f"original {''.join(i)}")
        for j in set_count:
            line[j] = '|'
        print(''.join(line))
        continue
    splitter_count = i.count('^')
    splitters = [i.index('^')]

    prev_count = len(set_count)

    last_splitter = splitters[0]
    for j in range(splitter_count-1):
        next_splitter = i.index('^', last_splitter+1)
        splitters.append(next_splitter)
        last_splitter = next_splitter
    #print(set_count)
    for j in splitters:
        if j not in set_count:
            continue
        splits += 1
        set_count.add(j-1)
        set_count.add(j+1)
        if j in set_count:
            set_count.remove(j)

    new_count = len(set_count)
    #print(set_count)
    line = i.copy()
    for j in set_count:
        # if line[j] == '^':
        #     continue
        line[j] = '|'
    print(''.join(line))
```
