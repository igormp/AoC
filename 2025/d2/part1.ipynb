```python
invalid_sum = 0
for i in inputs:
    start,end = i.split('-')
    start,end = int(start),int(end)
    for j in range(start,end+1):
        size = len(str(j))
        if size%2 != 0:
            continue
        if all([i == str(j)[0] for i in list(str(j))]):
            print(j)
            invalid_sum += j
            continue
        if str(j)[:size//2] == str(j)[size//2:]:
            print(j)
            invalid_sum += j
            continue
        # overkill, need to only check if repeated twice
        # for k in range(2,(size//2)+1):
        #     if size%k != 0:
        #         continue
        #     pieces = [str(j)[i:i+k] for i in range(0,size,k)]
        #     if all([i == pieces[0] for i in pieces]):
        #         print(j)
        #         invalid_sum += j
        #         break
```
