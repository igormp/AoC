```python
#FILE = "example.txt"
FILE = "input.txt"
```

```python
inputs = list(map(
    list, zip(*[map(str, line.split()) for line in open(FILE)])
))[0]
```

```python
op = {'L': -1, 'R': 1}
```

```python
parsed_inputs = [(op[i[0]] * int(i[1:])) for i in list(inputs)]
```

```python
cur_dial = 50
```

```python
def normalize_dial(dial_value):
    while dial_value > 99 or dial_value < 0:
        if dial_value > 99:
            dial_value = dial_value - 100
        elif dial_value < 0:
            dial_value = 100 - abs(dial_value)
    return dial_value
```

```python
zero_counts = 0
```

```python
for i in parsed_inputs:
    cur_dial = normalize_dial(cur_dial + i)
    # print(cur_dial)
    if cur_dial == 0:
        zero_counts += 1
```
