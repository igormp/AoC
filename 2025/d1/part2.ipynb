```python
#FILE = "example.txt"
FILE = "input.txt"
```

```python
inputs = list(map(
    list, zip(*[map(str, line.split()) for line in open(FILE)])
))[0]
```

```python
op = {'L': -1, 'R': 1}
```

```python
parsed_inputs = [(op[i[0]] * int(i[1:])) for i in list(inputs)]
```

```python
def normalize_dial(dial_value, move):
    ticks = 0 # if dial_value != 0 else -1
    distance = abs(move)%100
    ticks += abs(move) // 100
    dial_value += distance * (1 if move>=0 else -1)
    if ((dial_value <=0 and distance > abs(dial_value)) or dial_value>=100) and distance !=0:
        ticks+=1
    dial_value %= 100
    # if ticks > 0 and prev_dial == 0:
    #     ticks-=1
    return dial_value, ticks #(ticks if ticks > 0 else 0)
```

```python
cur_dial = 50
zero_counts = 0

#for i in [0, 50, 1000]:
for i in parsed_inputs:
    prev_dial = cur_dial
    cur_dial, ticks = normalize_dial(cur_dial, i)
    #print(f"start: {prev_dial:<3} move: {i:<6} end: {cur_dial:<3} ticks: {ticks:<3} is_zero: {cur_dial==0}")
    zero_counts += ticks
    # if cur_dial == 0:
    #     zero_counts += 1

print(zero_counts)
```
