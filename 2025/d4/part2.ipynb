```python
total_removed = 0
iter_count = 1
while True:
    movable_count = 0
    removed_positions = []
    for idx_i,row in enumerate(inputs):
        for idx_j,col in enumerate(row):
            if inputs[idx_i][idx_j] == '.':
                #print(f"{inputs[idx_i][idx_j]}", sep="", end="")
                continue
            near_rolls = 0
            neighbors = [(di,dj) for di in (-1,0,1) for dj in (-1,0,1) if not (di == 0 and dj == 0)]
            for di,dj in neighbors:
                ni, nj = idx_i+di, idx_j+dj
                if not (0 <= ni < len(inputs) and 0 <= nj < len(inputs[0])):
                    continue
                if inputs[ni][nj] == '@':
                    near_rolls += 1
            if near_rolls < 4:
                movable_count += 1
                removed_positions.append((idx_i, idx_j))
                #print("X", sep="", end="")
            else:
                pass
                #print(f"{inputs[idx_i][idx_j]}", sep="", end="")
        #print()
    for pos in removed_positions:
        inputs[pos[0]][pos[1]] = '.'
    print(f"{iter_count}: {movable_count}")
    iter_count += 1
    total_removed += movable_count
    if movable_count == 0:
        break
```
