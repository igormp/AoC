```python
import heapq
from collections import Counter, defaultdict
from itertools import product
from math import prod
import numpy as np
from dataclasses import dataclass
from functools import lru_cache
```

```python
#FILE = "example.txt"
FILE = "input.txt"
```

```python
inputs = [
    line.rstrip('\n') for line in open(FILE) if line.rstrip('\n') != ''
]
```

```python
len(inputs)
```

```python
presents = []
for i in range(1, 6*4, 4):
    area = 9- ''.join(inputs[i:i+3]).count('.')
    presents.append((inputs[i:i+3], area))
print(presents)
```

```python
class Tree:
    dims: tuple[int, int]
    area: int
    quantities: list[int]
    def __init__(self, dims: tuple[int, int], quantities: list[int]):
        self.dims = dims
        self.area = dims[0] * dims[1]
        self.quantities = quantities

trees: list[Tree] = []
for line in inputs[24:]:
    arr = line.split(' ')
    dims = tuple(map(int, arr[0].rstrip(':').split('x')))
    quantities = list(map(int, arr[1:]))
    trees.append(Tree(dims, quantities))
```

```python
fit_count = 0
for tree in trees:
    required_area = sum([presents[idx][1]*qnt for idx, qnt in enumerate(tree.quantities)])
    if required_area <= tree.area:
        fit_count += 1

    # print(tree.dims, tree.area, required_area)
fit_count
```
